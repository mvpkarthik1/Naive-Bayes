```python
#importing the required libraries

import pandas as pd
```

```python
# uploading the training dataset to train the model

SalaryData_Train = pd.read_csv("SalaryData_Train.csv")
```

```python
SalaryData_Train
```

```python
#finding the duplicated rows

SalaryData_Train[SalaryData_Train.duplicated()]
```

```python
#removing the duplicated rows

SalaryData_Train.drop_duplicates()
```

```python
# importing the label encoder from sklearn library, to label the dataset

from sklearn.preprocessing import LabelEncoder

le= LabelEncoder()
```

```python
categ=['workclass','education','maritalstatus','occupation','relationship','race','sex','Salary','native']

SalaryData_Train[categ]= SalaryData_Train[categ].apply(le.fit_transform)
```

```python
#plotting the heatmap for training dataset with the help of correlation....

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(20,10))
ax= plt.axes()
sns.heatmap(SalaryData_Train.corr(), annot= True, ax= ax)

ax.set_title(" TRAIN DATA HEAT MAP")
plt.show()
```

```python
#now importing the testing dataset for testing the model which is build by training with training dataset...

SalaryData_Test=pd.read_csv("SalaryData_Test.csv")
SalaryData_Test
```

```python
#finding the duplicated rows

SalaryData_Test[SalaryData_Test.duplicated()]
```

```python
#removing the duplicated rows..

SalaryData_Test.drop_duplicates()
```

```python
categ=['workclass','education','maritalstatus','occupation','relationship','race','sex','Salary','native']

SalaryData_Test[categ]=SalaryData_Test[categ].apply(le.fit_transform)
```

```python
SalaryData_Test
```

```python
# plotting the heat map for testing dataset

plt.figure(figsize=(20,10))

ax= plt.axes()
sns.heatmap(SalaryData_Test.corr(), annot= True, ax= ax)

ax.set_title("TEST DATA HEAT MAP")
plt.show()
```

```python
tr= SalaryData_Train.values
x_train= tr[:,:-1]
y_train= tr[:,-1]
```

```python
tx= SalaryData_Test.values
x_test= tx[:,:-1]
y_test= tx[:,-1]
```

```python
from sklearn.naive_bayes import MultinomialNB as NB
naive=NB()
```

```python
train_naive= naive.fit(x_train,y_train)
```

```python
model_train= train_naive.predict(x_train)
```

```python
import numpy as np

train_efficiency= np.mean(y_train==model_train)*100
```

```python
train_efficiency
```

```python
model_test= train_naive.predict(x_test)
```

```python
test_efficiency=np.mean(y_test==model_test)*100
```
